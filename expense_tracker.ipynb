```python
df['Month'] = df['Date'].dt.strftime('%b')
monthly_totals = df.groupby('Month')['Amount'].sum().reindex(
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
).dropna()
```

```python
monthly_totals
```

```python
plt.figure(figsize=(8,5))
monthly_totals.plot(kind='bar', color='teal')

plt.title('Monthly Total Revenue', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
monthly_expenses = df.groupby(['Month','Category'])['Amount'].sum().unstack(fill_value=0)
```

```python
monthly_expenses['Total'] = monthly_expenses.sum(axis=1)
monthly_expenses
```

```python
df['Category'].unique()
```

```python
colors = ['#a8dadc','#f4a261','#e9c46a','#90be6d','#f08080','#b5ead7']

# Create pie charts
months = monthly_expenses.index
plt.figure(figsize=(40,20))

for i, month in enumerate(months, 1):
    plt.subplot(1, len(months), i)
    plt.pie(monthly_expenses.loc[month],
            labels=monthly_expenses.columns,
            autopct='%1.1f%%',
            startangle=90,
            colors=colors)
    plt.title(f'{month} Expenses')

plt.show()
```

```python
monthly = df.groupby(['Month','Category'])['Amount'].sum().unstack(fill_value=0)
colors = ['#a8dadc','#f4a261','#e9c46a','#90be6d','#f08080','#b5ead7']

plt.figure(figsize=(10,8))

for i, month in enumerate(monthly.index):
    plt.subplot(2, (len(monthly.index)+1)//2, i+1)
    values = monthly.loc[month]

    wedges, texts = plt.pie(values,
                            startangle=90,
                            colors=colors)

    # Calculate percentages for legend
    total = values.sum()
    labels = [f"{cat}: {val/total*100:.1f}%" for cat, val in zip(values.index, values)]

    plt.legend(wedges, labels, title="Category", loc="best")
    plt.title(f'{month} Expenses')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))

for i, month in enumerate(monthly.index):
    plt.subplot(2, (len(monthly.index)+1)//2, i+1)
    values = monthly.loc[month]
    wedges, texts = plt.pie(values, startangle=90, colors=colors)


    total_spent = values.sum()
    top_category = values.idxmax()
    top_value = values.max()


    plt.text(1.3, 0.5, f"Total: {total_spent}", fontsize=11, weight='bold')
    plt.text(1.3, 0.3, f"Top: {top_category} ({top_value})", fontsize=11, weight='bold')

    plt.title(f'{month} Expenses')

plt.tight_layout()
plt.show()
```
